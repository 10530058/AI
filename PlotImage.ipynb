```python
# plot image
import matplotlib.pyplot as plt
from matplotlib.image import imread
img = imread('https://upload.wikimedia.org/wikipedia/commons/thumb/c/cf/Adobe_Photoshop_Express_logo.svg/1024px-Adobe_Photoshop_Express_logo.svg.png') 
plt.imshow(img)
plt.show()
```
