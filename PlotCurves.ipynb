```python
# coding: utf-8import 
import numpy as np
import matplotlib.pyplot as plt
class Matlab:
    """functions of class """
    def __init__(self, x , y):
        self.x = np.arange(x, y, 0.1)
        print("Initilized!")
    def plotSin(self):
        self.y = np.sin(self.x)
        plt.plot(self.x,self.y)
        plt.show()
    def plotCos(self):
        self.y = np.cos(self.x)
        plt.plot(self.x,self.y)
        plt.show()
```

```python
X=Matlab(1,6)
```

```python
X.plotSin()
```

```python
X.plotCos()
```
